# 3-1-(1). KMeans 학습

```python
from sklearn.preprocessing import scale
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
train_copy = train_df.copy().drop(columns = ['ID'], axis = 1)
```

```python
train_copy  # train_df에서 ID 피처만 제거한 것
```

```python
kmeans = KMeans(n_clusters = 2, init = 'k-means++', max_iter = 300, random_state = 0)
```

```python
# 전처리 하지 않은 기본 train_df 넣어서 kmeans(n_clusters = 2, init = 'k-means++', max_iter = 300, random_state = 0)
kmeans.fit(train_copy)
kmean_train = kmeans.transform(train_copy)
```

```python
kmean_train # train data transform 결과
```

```python
pd.DataFrame(kmeans.labels_).value_counts()
```

```python
print(kmeans.labels_)
```

```python
# 축
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)

# 점들을 데이터 프레임 형테로 저장
kmean_train_df = pd.DataFrame(kmean_train, columns=['X', 'Y'])
display(kmean_train_df.head())

# 점 데이터를 X-Y grid에 시각화하기
ax.scatter(kmean_train_df['X'], kmean_train_df['Y'], c='black', label='train data')


ax.set_xlabel('X') #x축 라벨
ax.set_ylabel('Y')#y축 라벨
ax.legend()
ax.grid()
```

```python
label_0 = kmean_train[kmeans.labels_ == 0]
label_1 = kmean_train[kmeans.labels_ == 1]
```

```python
label_0_df = pd.DataFrame(label_0, columns=['X', 'Y'])
label_1_df = pd.DataFrame(label_1, columns=['X', 'Y'])
```

```python
plt.scatter(label_0_df['X'], label_0_df['Y'], color = 'green', label = "0")
plt.scatter(label_1_df['X'], label_1_df['Y'], color = 'red', label = "1")
```
